### Task: Student Records — CRUD with Lists & Dicts

### Task Details: 
Implement a student record manager that stores records as dictionaries inside a list. Support add, view (by id), update, delete, and list all. Validate fields (age numeric, GPA 0.0–4.0).

```python
class StudentManager:
    def __init__(self):
        self.students = []
        self.next_id = 1000
    
    def add_student(self, name, age, gpa):
        if not name:
            return "Error: Name required"
        
        try:
            age = int(age)
            if age < 0 or age > 120:
                return "Error: Age must be 0-120"
        except:
            return "Error: Age must be integer"
        
        try:
            gpa = float(gpa)
            if gpa < 0 or gpa > 4.0:
                return "Error: GPA must be 0.0-4.0"
        except:
            return "Error: GPA must be number"
        
        student = {
            'id': self.next_id,
            'name': name,
            'age': age,
            'gpa': round(gpa, 2)
        }
        
        self.students.append(student)
        self.next_id += 1
        return f"Added student ID {student['id']}"
    
    def view_student(self, student_id):
        for student in self.students:
            if student['id'] == student_id:
                return student
        return "Student not found"
    
    def update_student(self, student_id, name=None, age=None, gpa=None):
        for student in self.students:
            if student['id'] == student_id:
                if name:
                    student['name'] = name
                if age:
                    try:
                        student['age'] = int(age)
                    except:
                        return "Error: Invalid age"
                if gpa:
                    try:
                        gpa_val = float(gpa)
                        if 0 <= gpa_val <= 4.0:
                            student['gpa'] = round(gpa_val, 2)
                        else:
                            return "Error: GPA out of range"
                    except:
                        return "Error: Invalid GPA"
                return "Updated successfully"
        return "Student not found"
    
    def delete_student(self, student_id):
        for i, student in enumerate(self.students):
            if student['id'] == student_id:
                self.students.pop(i)
                return "Deleted successfully"
        return "Student not found"
    
    def list_all(self):
        if not self.students:
            print("No students")
            return
        
        print("ID\tName\t\tAge\tGPA")
        print("-" * 40)
        for student in self.students:
            print(f"{student['id']}\t{student['name'][:15]}\t{student['age']}\t{student['gpa']}")


print("Sample usage:")
manager = StudentManager()
manager.add_student("Mubby isah", "20", "3.8")
manager.add_student("Maryam fatima", "22", "3.2")
manager.add_student("John joen", "19", "2.9")
manager.add_student("Adline grace", "21", "3.9")

manager.list_all()
manager.view_student(1000)  # Should show Alice
#manager.view_student(9999)  # Should show error

#manager.update_student(1001, gpa="3.5")  # Update Bob's GPA
#manager.update_student(1002, age="20", name="Charles Brown")  # Update Charlie
```
