### Task: Robust Calculator (types & conversions, error handling)



### Task Details: 
Write a calculator program that accepts two operands and an operator (+, -, *, /). Accept numbers as integers or floats (or numeric strings), convert types safely, and handle division-by-zero and invalid input with clear error messages. Add a mode that reads expressions like '12.5 * 3' and evaluates them after validating tokens.

```python
def simple_calculator():
    """Basic calculator with error handling."""
    print("Calculator - Choose mode: 'expression' or 'manual'")
    mode = input("Mode: ").lower()
    
    if mode == 'expression':
        return evaluate_expression()
    else:
        return manual_mode()

def manual_mode():
    """Get inputs separately."""
    try:
        num1_input = input("First number: ")
        num1 = float(num1_input) if '.' in num1_input else int(num1_input)
        
        operator = input("Operator (+, -, *, /): ")
        
        num2_input = input("Second number: ")
        num2 = float(num2_input) if '.' in num2_input else int(num2_input)
        
        if operator == '+':
            result = num1 + num2
        elif operator == '-':
            result = num1 - num2
        elif operator == '*':
            result = num1 * num2
        elif operator == '/':
            if num2 == 0:
                return "Error: Cannot divide by zero"
            result = num1 / num2
        else:
            return "Error: Invalid operator"
        
        return f"Result: {result}"
        
    except ValueError:
        return "Error: Invalid number format"
    except Exception as e:
        return f"Error: {e}"

def evaluate_expression():
    """Parse expressions like '12.5 * 3'."""
    expr = input("Enter expression (num op num): ")
    
    try:
        parts = expr.split()
        if len(parts) != 3:
            return "Error: Need 3 parts (num op num)"
        
        num1 = float(parts[0]) if '.' in parts[0] else int(parts[0])
        op = parts[1]
        num2 = float(parts[2]) if '.' in parts[2] else int(parts[2])
        
        if op == '+':
            result = num1 + num2
        elif op == '-':
            result = num1 - num2
        elif op == '*':
            result = num1 * num2
        elif op == '/':
            if num2 == 0:
                return "Error: Division by zero"
            result = num1 / num2
        else:
            return "Error: Invalid operator"
        
        return f"Result: {result}"
        
    except ValueError:
        return "Error: Invalid numbers"
    except Exception as e:
        return f"Error: {e}"

Test = 100+50
print(f"Answer (test question is 100+50): {Test}")
```
