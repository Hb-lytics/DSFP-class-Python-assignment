### Task: Loop Practice — Prime Gap Finder (loops)

### Task Details: 
Writescript to find all pairs of consecutive primes ≤ 10,000 with gaps ≥ 20 and save them

```python
def is_prime(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    
    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
    return True

def find_prime_gaps(limit=10000, min_gap=20):
    primes = []
    gaps = []
    
    # Find all primes up to limit
    for num in range(2, limit + 1):
        if is_prime(num):
            primes.append(num)
    
    # Check consecutive gaps
    for i in range(len(primes) - 1):
        gap = primes[i + 1] - primes[i]
        if gap >= min_gap:
            gaps.append((primes[i], primes[i + 1], gap))
    
    return gaps

def save_gaps_to_file(gaps, filename="prime_gaps.txt"):
    try:
        with open(filename, 'w') as f:
            f.write("Prime pairs with gap >= 20\n")
            f.write("=" * 40 + "\n")
            for p1, p2, gap in gaps:
                f.write(f"{p1} -> {p2} (gap: {gap})\n")
        
        print(f"Saved {len(gaps)} gaps to {filename}")
        return True
    except Exception as e:
        print(f"Error saving: {e}")
        return False

print("Finding prime gaps...")
gaps = find_prime_gaps(10000, 20)
print(f"Found {len(gaps)} prime pairs with gap >= 20")
if gaps:
    print("First few gaps:")
    for i in range(min(5, len(gaps))):
        p1, p2, gap = gaps[i]
        print(f"  {p1} -> {p2} (gap: {gap})")

save_gaps_to_file(gaps)
```
