### Task Details: 
Given students assessment scores, compute weighted totals and assign letter grades with +/- tiers using nested conditionals. Produce class summary statistics (mean, median, grade distribution) and write a short text report.

```python
def calculate_grades(student_data):
    results = []
    
    for student in student_data:
        hw = student.get('homework', 0)
        mid = student.get('midterm', 0)
        final = student.get('exam', 0)
        
        total = hw * 0.1 + mid * 0.2 + final * 0.7
        total = round(total, 1)
        
        # Grade assignment with nested conditionals
        if total >= 70:
            grade = "A"
        elif total >= 60:
            grade = "B"
        elif total >= 50:   
            grade = "C"
        elif total >= 30:
            grade = "D"
        else:
            grade = "F"
        
        results.append({
            'name': student['name'],
            'total': total,
            'grade': grade
        })
    
    return results

def calculate_statistics(grade_results):
    if not grade_results:
        return {}
    
    scores = [s['total'] for s in grade_results]
    mean = sum(scores) / len(scores)
    
    sorted_scores = sorted(scores)
    n = len(sorted_scores)
    if n % 2 == 0:
        median = (sorted_scores[n//2 - 1] + sorted_scores[n//2]) / 2
    else:
        median = sorted_scores[n//2]
    
    grade_dist = {}
    for result in grade_results:
        grade = result['grade'][0]
        grade_dist[grade] = grade_dist.get(grade, 0) + 1
    
    return {
        'mean': round(mean, 2),
        'median': round(median, 2),
        'distribution': grade_dist,
        'count': len(grade_results)
    }


print("Sample:")
sample_data = [
    {'name': 'Mubby', 'homework': 95, 'midterm': 88, 'exam': 92},
    {'name': 'Maryam', 'homework': 78, 'midterm': 82, 'exam': 75}
]
grades = calculate_grades(sample_data)
stats = calculate_statistics(grades)
print("Grades:", grades)
print("Stats:", stats)
```
